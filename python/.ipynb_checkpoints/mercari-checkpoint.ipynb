```python
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
import numpy as np

from sklearn.base import TransformerMixin
class DataFrameImputer(TransformerMixin):
    def fit(self, X, y=None):
        self.fill = pd.Series([X[c].value_counts().index[0]
            if X[c].dtype == np.dtype('O') else X[c].median() for c in X],
            index=X.columns)
        return self
    def transform(self, X, y=None):
        return X.fillna(self.fill)
```

```python
#train_X = train_df.drop(['item_id','category_class','listing_at','item_tag_hash'],axis=1)

train_X = pd.get_dummies(train_df.drop('category_class',axis=1)) # å…¨éƒ¨ä¹—ã›ï¼Žã ã‘ã©ã“ã‚Œã¯ã©ã†ãªã‚“ã ã‚ã†ï¼Žç²¾åº¦ã¯è‰¯ããªã‚‹ãŒï¼Žï¼Žï¼Žï¼Ž
train_y = train_df['category_class']

# ãƒˆãƒ¬ã‚¤ãƒ³ãƒ‡ãƒ¼ã‚¿ã¨ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã«åˆ†ã‘ã‚‹
from sklearn.model_selection import train_test_split
(train_X, test_X ,train_y, test_y) = train_test_split(train_X, train_y, test_size = 0.3, random_state = 666)

gbm = xgb.XGBClassifier(max_depth=3, n_estimators=300, learning_rate=0.05).fit(train_X, train_y)
y_pred = gbm.predict(test_X)
```

```python
from sklearn.metrics import f1_score 
score = f1_score(y_pred, test_y, average='macro')
```

```python
import pandas as pd
train_df = pd.read_csv("./data/train.csv")
test_df = pd.read_csv("./data/test.csv")
```

```python
#train_X = train_df.drop(['category_class','listing_at','item_tag_hash'],axis=1)
#train_X = train_df.drop(['category_class','listing_at','item_tag_hash','area_name'],axis=1) #ã“ã£ã¡ã®æ–¹ãŒä¸Šã‚ˆã‚Šã¯è‰¯ã„
train_X = train_df.drop(['category_class','listing_at','item_tag_hash','item_id'],axis=1) # ã•ã‚‰ã«ä¸Šã‚ˆã‚Šã‚‚è‰¯ã„

train_X = pd.get_dummies(train_X)
train_y = train_df['category_class']
```

```python
# ãƒˆãƒ¬ã‚¤ãƒ³ãƒ‡ãƒ¼ã‚¿ã¨ãƒ†ã‚¹ãƒˆãƒ‡ãƒ¼ã‚¿ã«åˆ†ã‘ã‚‹
from sklearn.model_selection import train_test_split
(train_X, test_X ,train_y, test_y) = train_test_split(train_X, train_y, test_size = 0.3, random_state = 666)
```

```python
train_X
```

```python
train_y
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
clf = clf.fit(train_X, train_y)
pred = clf.predict(test_X)
```

```python
pred
```

```python
from sklearn.metrics import f1_score 
f1_score(pred, test_y, average='macro')
```
